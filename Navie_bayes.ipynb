```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('fourthClass_Navie_bayes.csv')
print(df)
```

```python
def run_naive_bayes_classification():
    
    print("--- Dataset ---")
    print(df)   
    
    X = df.drop('Species', axis=1)
    y = df['Species']
    
    encoder = OrdinalEncoder()
    X_encoded = encoder.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.25, random_state=42)
    
    model = CategoricalNB(alpha = 1.0)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nAccuracy: {accuracy * 100}%")
    
   
    new_data = pd.DataFrame([['white', 'three', 'short', 'yes']],columns = ['color', 'legs', 'Height', 'Smell'])
    
    new_encoded = encoder.transform(new_data)
    prediction = model.predict(new_encoded)
    
    
    
    print(f"Predictions: {y_pred}")
    print(f"Actual: {y_test.values}")

if __name__ == "__main__":
    run_naive_bayes_classification()
```
