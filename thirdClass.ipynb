```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Prepare the dataset
data = {
    'TV': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7],
    'Radio': [37.8, 39.3, 45.9, 41.3, 10.8, 48.9],
    'Newspaper': [69.2, 45.1, 69.3, 58.5, 58.4, 75.0],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2]
}

df = pd.DataFrame(data)

# 2. Define Features (X) and Target (y)
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

# 3. Initialize and Fit the Model
# Given the small size of this sample (6 rows), we'll fit on the whole set 
# or a simple split. Usually, you'd use train_test_split.
model = LinearRegression()
model.fit(X, y)

# 4. Make Predictions
predictions = model.predict(X)

# 5. Output Results
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}")
print("-" * 30)

# Show actual vs predicted
for actual, pred in zip(y, predictions):
    print(f"Actual Sales: {actual} | Predicted Sales: {pred:.2f}")

# 6. Model Evaluation
mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)
print("-" * 30)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

```python
# --- Cell 1: Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# --- Cell 2: Data Loading ---
# Creating the dataframe from your provided data
data = {
    'TV': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7],
    'Radio': [37.8, 39.3, 45.9, 41.3, 10.8, 48.9],
    'Newspaper': [69.2, 45.1, 69.3, 58.5, 58.4, 75.0],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2]
}
df = pd.DataFrame(data)

# In Jupyter, just typing the variable name at the end shows a nice table
df
```

```python
# --- Cell 3: Modeling ---
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

model = LinearRegression()
model.fit(X, y)

print(f"Model Training Complete.")
print(f"R-squared Score: {model.score(X, y):.4f}")
```

```python
# --- Cell 4: Visualization (The best part of Jupyter) ---
# This creates a visual check of Actual vs Predicted Sales
predictions = model.predict(X)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=y, y=predictions)
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r') # Diagonal line
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted')
plt.show()
```
